```python
#Add Pandas dependency 
import pandas as pd
import os
```

```python
#files to load
schoolDataToLoad = os.path.join("Resources", "schools_complete.csv")
studentDataToLoad = os.path.join("Resources","students_complete.csv")
```

```python
#Read school file and store in a Pandas dataframe
schoolData_df = pd.read_csv(schoolDataToLoad)
schoolData_df
```

```python
#Read student file and store in a Pandas dataframe
studentData_df = pd.read_csv(studentDataToLoad)
studentData_df
```

```python
#Add each prefix and suffix to a list
pre_suf = ["Dr. ", "Mr. ","Ms. ", "Mrs. ", "Miss ", " MD", " DDS", " DVM", " PhD"]

# Iterate through the words in the "prefixes_suffixes" list and replace them with an empty space, "".
for word in pre_suf:
    studentData_df["student_name"] = studentData_df["student_name"].str.replace(word,"")
```

```python
# Combine the data into a single dataset.
schoolDataComplete_df = pd.merge(studentData_df, schoolData_df, on=["school_name", "school_name"])
schoolDataComplete_df.head()
```

```python
#Get total number of students
studentCount = schoolDataComplete_df["Student ID"].count()
studentCount
```

```python
#Get total number of schools
schoolCount = schoolData_df["school_name"].count()
schoolCount
```

```python
#Get total budget
totalBudget = schoolData_df["budget"].sum()
totalBudget
```

```python
#Get average reading score
avgReadingScore = schoolDataComplete_df["reading_score"].mean()
avgReadingScore
```

```python
#get average math score
avgMathScore = schoolDataComplete_df["math_score"].mean()
avgMathScore
```

```python
#Get all passing students
passingMath = schoolDataComplete_df[schoolDataComplete_df["math_score"] >= 70]
passingMath.head()
passingReading = schoolDataComplete_df[schoolDataComplete_df["reading_score"] >= 70]
passingReading.head()
```

```python
#Calculate number of students passing math or reading
passingMathCount = passingMath["student_name"].count()
passingReadingCount = passingReading["student_name"].count()

print(passingMathCount)
print(passingReadingCount)
```

```python
#calculate percentages of students passing math or reading
passingMathPercent = passingMathCount / float(studentCount)* 100
passingReadingPercent = passingReadingCount / float(studentCount)* 100

print(passingMathPercent)
print(passingReadingPercent)
```

```python
#find students passing math AND reading
passingBoth = schoolDataComplete_df[(schoolDataComplete_df["math_score"] >= 70) & (schoolDataComplete_df["reading_score"] >= 70)]
passingBoth.head()
```

```python
#get total students passing both
passingBothCount = passingBoth["student_name"].count()
#get total percentage
passingBothPercent = passingBothCount / studentCount * 100
passingBothPercent
```

```python
# Adding a list of values with keys to create a new DataFrame.
districtSummary_df = pd.DataFrame(
          [{"Total Schools": schoolCount,
          "Total Students": studentCount,
          "Total Budget": totalBudget,
          "Average Math Score": avgMathScore,
          "Average Reading Score": avgReadingScore,
          "% Passing Math": passingMathPercent,
         "% Passing Reading": passingReadingPercent,
        "% Overall Passing": passingBothPercent}])

#Format columns
districtSummary_df["Total Students"] =districtSummary_df["Total Students"].map("{:,}".format)
districtSummary_df["Total Budget"] =districtSummary_df["Total Budget"].map("${:,.2f}".format)
districtSummary_df["Average Math Score"] = districtSummary_df["Average Math Score"].map("{:.1f}".format)
districtSummary_df["Average Reading Score"] = districtSummary_df["Average Reading Score"].map("{:.1f}".format)
districtSummary_df["% Passing Math"] = districtSummary_df["% Passing Math"].map("{:.0f}".format)
districtSummary_df["% Passing Reading"] = districtSummary_df["% Passing Reading"].map("{:.0f}".format)
districtSummary_df["% Overall Passing"] = districtSummary_df["% Overall Passing"].map("{:.0f}".format)

districtSummary_df
```

```python
# Reorder the columns in the order you want them to appear.
newOrder = ["Total Schools", "Total Students", "Total Budget","Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "% Overall Passing"]

# Assign district summary df the new column order.
districtSummary_df = districtSummary_df[newOrder]
districtSummary_df
```
